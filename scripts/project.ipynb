```python
import minsearch
import pandas as pd
from openai import OpenAI
```

```python
df = pd.read_csv('../data/100_Factual_Medical_Cases_Dataset.csv')
df
```

```python
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.lower()
df.columns
```

```python
df = df.drop_duplicates(subset=['disease_name'])
```

```python
df
```

```python
dataset = df.to_dict(orient='records')
dataset[0]
```

```python
index = minsearch.Index(
    text_fields=['diseased', 'disease_name', 'disease_type', 'prevalence',
       'transmission_mode', 'incubation_period', 'complications',
       'recommended_treatment', 'prevention_measures',
       'recurrence_probability', 'symptom_category', 'symptom_duration',
       'frequency', 'diagnosis_method', 'time_to_diagnosis', 'treatment_type',
       'medication_name', 'dosage', 'treatment_duration',
       'treatment_side_effects', 'symptoms'],
    keyword_fields=['diseased', 'disease_name', 'disease_type', 'prevalence',
       'transmission_mode', 'incubation_period', 'complications',
       'recommended_treatment', 'prevention_measures',
       'recurrence_probability', 'symptom_category', 'symptom_duration',
       'frequency', 'diagnosis_method', 'time_to_diagnosis', 'treatment_type',
       'medication_name', 'dosage', 'treatment_duration',
       'treatment_side_effects', 'symptoms']
)
```

```python
index.fit(dataset)
```

```python
def search(query):

    boost = {}
    
    results = index.search(
        query=query,
        filter_dict = {},
        boost_dict = boost,
        num_results=10
    )
    return results
```

```python
def build_prompt(query, search_results):
    context_template = """
        diseased: {diseased}
        disease_name: {disease_name}
        disease_type: {disease_type}
        prevalence: {prevalence}
        transmission_mode: {transmission_mode}
        incubation_period: {incubation_period}
        complications: {complications}
        recommended_treatment: {recommended_treatment}
        prevention_measures: {prevention_measures}
        recurrence_probability:{recurrence_probability}
        symptom_category: {symptom_category}
        symptom_duration: {symptom_duration}
        frequency: {frequency}
        diagnosis_method: {diagnosis_method}
        time_to_diagnosis: {time_to_diagnosis}
        treatment_type: {treatment_type}
        medication_name: {medication_name}
        dosage: {dosage}
        treatment_duration: {treatment_duration}
        treatment_side_effects: {treatment_side_effects}
        symptoms: {symptoms}
    """.strip()
    prompt_template = """
        You're a medical assistant. Answer the QUESTION based on the CONTEXT from the FAQ database.
        Use only the facts from the CONTEXT when answering the QUESTION.

        QUESTION: {question}

        CONTEXT: 
        {context}
    """.strip()

    context = ""
    
    for doc in search_results:
        context = context + context_template.format(**doc)+ "\n\n"
    
    prompt = prompt_template.format(question=query, context=context).strip()
    return prompt
```

```python
query = "Patient with malaria"
search_results = search(query)
prompt = build_prompt(query, search_results)
```
